**Задание 2**

Постройте график, который будет разбит на 4 части и в каждой части будут гистограммы для следующих перемменных:

- bill_length_mm
- bill_depth_mm
- flipper_length_mm
- body_mass_g



```python
# Комбинированный график  Matplotlib + Seaborn

penguins = sns.load_dataset("penguins").dropna()

fig, ax = plt.subplots(2, 2, figsize=(12, 8))  # 4 подграфика

ax[0, 0].hist(penguins["bill_length_mm"], bins=20, color="skyblue", edgecolor="black")
ax[0, 0].set_title("bill_length_mm - Matplotlib")
ax[0, 0].set_xlabel("Bill Length (mm)")
ax[0, 0].set_ylabel("Частота")

sns.histplot(penguins["bill_depth_mm"], bins=20, kde=True, color="lightgreen", ax=ax[0, 1])
ax[0, 1].set_title("bill_depth_mm - Seaborn")
ax[0, 1].set_xlabel("Bill Depth (mm)")
ax[0, 1].set_ylabel("Частота")

ax[1, 0].hist(penguins["flipper_length_mm"], bins=20, color="orange", edgecolor="black")
ax[1, 0].set_title("flipper_length_mm - Matplotlib")
ax[1, 0].set_xlabel("Flipper Length (mm)")
ax[1, 0].set_ylabel("Частота")


sns.histplot(penguins["body_mass_g"], bins=20, kde=True, color="lightcoral", ax=ax[1, 1])
ax[1, 1].set_title("body_mass_g - Seaborn")
ax[1, 1].set_xlabel("Mass (g)")
ax[1, 1].set_ylabel("Частота")

plt.tight_layout()
plt.show()
```

**Задание 3**

Постройте график body_mass_g в зависимости от species



```python
penguins = sns.load_dataset("penguins").dropna()

fig, ax = plt.subplots(1, 2, figsize=(10, 4.5))

# Boxplot
sns.boxplot(x="species", y="body_mass_g", hue="species", data=penguins, ax=ax[0], palette="Set2", legend=False)
ax[0].set_title("Зависимость между массой и видом")

# Violinplot
sns.violinplot(x="species", y="body_mass_g", hue="species", data=penguins, ax=ax[1], palette="Set2", legend=False)
ax[1].set_title("Зависимость между массой и видом")

plt.tight_layout()
plt.show()
```

```python
# Plotly
penguins = sns.load_dataset("penguins").dropna()
fig = px.histogram(penguins, x="body_mass_g", color="species", nbins=20, title="Зависимость между массой и видом пингвина")
fig.update_layout(width=600, height=400)
fig.show()
```
